```python
import numpy as np
#This code includes the periodic boundary condition also named PBC.
#The below code is changing the dimensions i.e., changing box width, height and length keeping others constant as given in the question.

#Prepare the cubic box with Neon atoms
a = 25.0         # box length in angstrom
b = 25.0         # box width
c = 25.0         # box height

N = 260        # number of particles to be used in the simulations
#Adding the boundary condition related algorithm here only.
def pbc_points(point,xyz):
  x_boxsize=25.0 #boxlength
  y_boxsize=25.0 #boxwidth
  z_boxsize=25.0 #boxheight
  N = len(xyz)
  #for N number of particle, distances from jth particle
  pbc_xyz=np.empty_like(xyz)
  pbc = True
  for i in range(N):
    if pbc :
      dx = point[0,0]  - xyz[i,0] #distance between particle  j and i in x direction
      if dx > x_boxsize * 0.5 : dx = dx - x_boxsize
      if dx <= -x_boxsize * 0.5 : dx = dx + x_boxsize

      dy = point[0,1] - xyz[i,1] #distance between particle  j and i in y direction
      if dy > y_boxsize * 0.5 : dy = dy - y_boxsize
      if dy <= -y_boxsize * 0.5 : dy = dy + y_boxsize

      dz = point[0,2] - xyz[i,2] #distance between particle  j and i in z direction
      if dz > z_boxsize * 0.5 : dz = dz - z_boxsize
      if dz <= -z_boxsize * 0.5 : dz = dz + z_boxsize

      pbc_xyz[i,0]= point[0,0] - dx
      pbc_xyz[i,1]= point[0,1] - dy
      pbc_xyz[i,2]= point[0,2] - dz
  return pbc_xyz

def distance(p, points, min_distance):
    """
    Determines if any points in the list are less than the minimum specified
    distance apart from all others.
    p: point (cartesian coordinates)
    points: all other points (cartesian coordinates)
    """
    distances = np.empty(0, float)
    for i in range(len(points)) :
       pxyz=pbc_points(p,points)
       #calculate all distances
       distances = np.append(distances,np.sqrt(np.sum(np.square(np.subtract(p[0,:],pxyz[i,:])))))

    distances = np.where(distances < min_distance)
    #check all distances are less than the minimum distances

    return distances[0].size < 1
    #return only if all the distances are greater than the minimum

points = np.empty((0, 3), float)       # x, y, z columnwise
min_distance=3.2 #in angstrom

###### draw the Neon atom positions randomly in the box
#origin of the box is at 0,0,0
import random
while len(points) < N:
    x = random.uniform(0, a) #sample a real number from 0 to a (length of the box)
    y = random.uniform(0, b)
    z = random.uniform(0, c)
    p = np.array([[x,y,z]])
    if len(points) == 0:                # add first point blindly #This the random atom we are choosing
         points=np.append(points,p,axis=0)
    elif distance(p, points, min_distance):     # ensure the minimum distance is met
        points = np.vstack((points, p))
        print(len(points),p)

init_pos=np.zeros((N,3))
init_pos[:,:]=points[:,:]
```

```python
from scipy.constants import Boltzmann #Boltzmann constant in Joule/Kelvin
mass_of_Neon=20.1797*1822.89 #in atomic unit

N=260
def init_vel(N,mass,temp):
   rn = np.random.normal(0,1,size=N) #Sample from a normal ditribution
   #temp: temperature
   #mass: mass of the particle
   #sqrt(KbT/m)
   return rn * np.sqrt((Boltzmann / 1.602e-19) *  temp/ mass)#converted into (eV*s/angstrom/mass_au)

temp=300 #T=300 K
vx=init_vel(N,mass_of_Neon,temp) #Velocity along x axis
vy=init_vel(N,mass_of_Neon,temp) #along y axis
vz=init_vel(N,mass_of_Neon,temp) #along z axis

vx = vx - np.sum(vx)/N
vy = vy - np.sum(vy)/N
vz = vz - np.sum(vz)/N

init_vel=np.transpose(np.vstack([vx,vy,vz]))
print(init_vel)
```

```python
def Ne_Ne_LJ_pot(r):
  epsilon=0.0030840 #maximum depth of the potential in eV
  sigma=2.801 #equilibrium distance in angstrom
  return 4*epsilon*((sigma/r)**12 - (sigma/r)**6)

def Ne_Ne_LJ_force(r): #-dV/dr with repect to distance
  epsilon=0.0030840 #maximum depth of the potential in eV
  sigma=2.801 #equilibrium distance in angstrom
  return 48 * epsilon * np.power(sigma, 12) / np.power(r, 13) - 24 * epsilon \
         * np.power(sigma, 6) / np.power(r, 7)

def force_xyz(pos1,pos2): #dV/dx = dV/dr*dr/dx  with respect to cartesian coordinates
    dist=np.sqrt(np.sum((np.subtract(pos1,pos2))**2))
    fr=Ne_Ne_LJ_force(dist)
    force1=fr/dist*np.subtract(pos1,pos2)
    force2=-fr/dist*np.subtract(pos1,pos2)
    return force1, force2
```

```python
def box_pbc(xyz):
  x_boxsize=25.0 #boxlength
  y_boxsize=25.0 #boxwidth
  z_boxsize=25.0 #boxheight

  N=260
  #origin at 0,0
  #for N number of particle, distances from jth particle
  pbc_xyz=np.empty_like(xyz)
  pbc = True
  for i in range(N):
    if pbc :
      dx = xyz[i,0] #distance of particle  i from origin in x direction
      if dx > x_boxsize : dx = dx - x_boxsize
      if dx <= 0 : dx = dx + x_boxsize

      dy = xyz[i,1] #distance of particle i from origin in y direction
      if dy > y_boxsize : dy = dy - y_boxsize
      if dy <= 0 : dy = dy + y_boxsize

      dz = xyz[i,2] #distance of particle i from origin in z direction
      if dz > z_boxsize : dz = dz - z_boxsize
      if dz <= 0 : dz = dz + z_boxsize

      pbc_xyz[i,0]= dx
      pbc_xyz[i,1]= dy
      pbc_xyz[i,2]= dz


  return pbc_xyz
```

```python
def calc_acc(pos_t):
  N=260
  mass_of_Neon=20.1797*1822.39
  acc_t=np.zeros((N,3))
  for i in range(N):
    pxyz=pbc(pos_t,i)
    for j in range(N):
      if j > i :
        f1, f2 =force_xyz(pxyz[i,:],pxyz[j,:])
        acc_t[i,:] += f1/mass_of_Neon
        acc_t[j,:] += f2/mass_of_Neon
  return acc_t
```

```python
def com_vel_remove(vel):
    N = len(vel)
    vel[:,0] = vel[:,0] - np.sum(vel[:,0])/N
    vel[:,1] = vel[:,1] - np.sum(vel[:,1])/N
    vel[:,2] = vel[:,2] - np.sum(vel[:,2])/N
    return vel
```

```python
def velocity_verlet1 (old_pos, old_vel, old_acc, dt):
  #step 1
  vhalf = old_vel + 0.5*old_acc*dt

  #step 2
  new_pos = old_pos + vhalf*dt
  new_pos = box_pbc(new_pos)
  #step 3
  new_acc = calc_acc(new_pos)

  #step 4
  new_vel = vhalf + 0.5*new_acc*dt
  new_vel = com_vel_remove(new_vel)

  return new_pos, new_vel, new_acc
```

```python
def velocity_verlet (old_pos, old_vel, old_acc, dt):
  #step 1
  new_pos = old_pos + old_vel*dt + 0.5*old_acc*dt**2
  new_pos = box_pbc(new_pos)
  #step 2
  new_acc = calc_acc(new_pos)

  #step 3
  new_vel = old_vel + 0.5*(old_acc+new_acc)*dt
  new_vel = com_vel_remove(new_vel)

  return new_pos, new_vel, new_acc
```

```python
def vel_rescale(ref_temp, velocity):
    N=260
    mass_of_Neon=20.1797*1822.39
    kb=8.6182833957553e-05
    vel=np.sqrt(np.square(velocity[:,0])+np.square(velocity[:,1])+np.square(velocity[:,2]))
    ke=np.sum(0.5*mass_of_Neon*vel**2)
    #temperature of the molecules (system)
    temp=ke/1.5/kb/N
    #scale factor
    scale = np.sqrt(ref_temp/temp)
    velocity = velocity * scale
    return velocity
```

```python
def total_ener_temp(positions, velocity):
  N=260
  mass_of_Neon=20.1797*1822.39
  kb=8.6182833957553e-05
  poten=0.0
  for i in range(N):
    for j in range(i+1,N):
     dist=np.sqrt(np.sum(np.square(np.subtract(positions[i,:],positions[j,:]))))
     poten += Ne_Ne_LJ_pot(dist)

  vel=np.sqrt(np.square(velocity[:,0])+np.square(velocity[:,1])+np.square(velocity[:,2]))
  ke=np.sum(0.5*mass_of_Neon*vel**2)
  return poten, ke,  ke/1.5/kb/260
```
